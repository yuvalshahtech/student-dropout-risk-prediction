```python
import pandas as pd

df = pd.read_csv("../data/sample_students.csv")
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6,4))
sns.heatmap(df.isnull())
plt.title("Missingness Heatmap")
plt.savefig("../results/eda/missingness.png")
plt.show()
```

```python
plt.hist(df["age"], bins=20, color="orange", edgecolor="black")
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("No. of Students")
plt.savefig("../results/eda/age_hist.png")
plt.show()
```

```python
plt.hist(df["attendance_pct"], bins=20, color="green", edgecolor="black")
plt.title("Attendance Distribution")
plt.xlabel("% of Attendance")
plt.ylabel("No. of Students")
plt.savefig("../results/eda/attendance_hist.png")
plt.show()
```

```python
import json
import numpy as np

df["assignment_mean"] = df["assignment_scores"].apply(lambda x: np.mean(json.loads(x)))
plt.hist(df["assignment_mean"], bins=20, color="blue", edgecolor="black")
plt.title("Assignment Score Mean Distribution")
plt.xlabel("Mean Score")
plt.ylabel("No. of Students")
plt.savefig("../results/eda/assignment_mean_hist.png")
plt.show()
```

```python
gender_counts = df["gender"].value_counts()

plt.bar(gender_counts.index, gender_counts.values, color="purple", edgecolor="black")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.savefig("../results/eda/gender_count.png")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Matrix")
plt.savefig("../results/eda/corr.png")
plt.show()
```

```python
df['final_grade'].value_counts().plot(kind='bar', color="skyblue", edgecolor="black")
plt.title("Final Grade Distribution")
plt.xlabel("Grades")
plt.ylabel("No. of Students")
plt.savefig("../results/eda/final_grade_balance.png")
plt.show()
```
